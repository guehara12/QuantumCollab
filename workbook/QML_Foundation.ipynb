# Mathematical Foundations of Variational Quantum Circuits

## The Mathematical Language of QML
1. Linear Algebra Fundamentals
In QML, a qubit is a vector in a 2-dimensional complex vector space (Hilbert Space). We use **Bra-Ket** notation to represent these vectors.
* The Ket (State Vector): $|0\rangle = \begin{pmatrix} 1 \\ 0 \end{pmatrix}$ and $|1\rangle = \begin{pmatrix} 0 \\ 1 \end{pmatrix}$.
* Superposition: Any state $|\psi\rangle$ is a linear combination: $|\psi\rangle = \alpha|0\rangle + \beta|1\rangle$, where $\alpha, \beta \in \mathbb{C}$ and $|\alpha|^2 + |\beta|^2 = 1$.
* Inner Product (The "Overlap"): $\langle \psi | \phi \rangle$. If this is 0, the states are orthogonal.

## 2. Unitary Operators (The "Rules of Motion")
Every quantum gate is a Unitary Matrix $U$.
* Definition: $U^\dagger U = I$, where $U^\dagger$ is the conjugate transpose.
* Preservation of Norm: Because $U$ is unitary, the "length" (probability) of the state vector is always 1. Applying a gate is essentially rotating the vector in Hilbert space without stretching or shrinking it.

## 3. The Pauli Gates (The Building Blocks)
The Pauli matrices $\{X, Y, Z\}$ form a basis for the space of $2 \times 2$ Hermitian matrices. In QML, they are used both as gates (to move the state) and observables (to measure the state).

### The X-Gate (The Quantum NOT)
$$X = \begin{pmatrix} 0 & 1 \\ 1 & 0 \end{pmatrix}$$Action: Flips $|0\rangle$ to $|1\rangle$ and vice versa. It is a rotation of $\pi$ radians around the X-axis of the Bloch sphere.

### The Z-Gate (The Phase Flip)
$$Z = \begin{pmatrix} 1 & 0 \\ 0 & -1 \end{pmatrix}$$Action: Leaves $|0\rangle$ unchanged but flips the sign (phase) of $|1\rangle$.Math: $Z|0\rangle = |0\rangle$, $Z|1\rangle = -|1\rangle$.
### The Y-Gate
$$Y = \begin{pmatrix} 0 & -i \\ i & 0 \end{pmatrix}$$Action: Combines a bit flip and a phase flip.

## I. Introduction: The QML Paradigm
In classical machine learning, we map data to a feature space and optimize a function. In Quantum Machine Learning (QML), we map data into a Hilbert Space—a complex vector space of $2^n$ dimensions—and optimize a Unitary Transformation.

### The Three-Stage Pipeline
* State Preparation ($\Phi$): Classical Data $\to$ Quantum State.
* Variational Ansatz ($U_\theta$): Trainable weight layers.
* Measurement ($\hat{M}$): Quantum State $\to$ Classical Prediction.

## II. Stage 1: State Preparation (The Feature Map)
State preparation is the process of encoding classical data $x$ into a quantum state $|\psi(x)\rangle$. In QML, this is often achieved through a **Feature Map**, which is a non-linear transformation that maps data into a high-dimensional Hilbert space.Mathematically, we apply a unitary transformation $U_{\Phi}(x)$ to the initial state $|0\rangle^{\otimes n}$:$$|\psi(x)\rangle = U_{\Phi}(x)|0\rangle^{\otimes n}$$A common method is Angle Encoding, where each feature $x_i$ is encoded as a rotation gate:$$U_{\Phi}(x) = \bigotimes_{i=1}^{n} R_y(x_i)$$For more complex data, we use the Z-Feature Map or ZZ-Feature Map, which introduces entanglement to capture correlations between features:$$U_{\Phi}(x) = \exp\left(i \sum_{S \subseteq \{1, \dots, n\}} \phi_S(x) \prod_{j \in S} Z_j\right)$$

We can think about this too, to process classical data $x \in \mathbb{R}^n$, we must encode it into a quantum state $|\psi(x)\rangle$.

### The Mathematics of Angle Encoding
We represent each feature $x_i$ as a rotation angle for a specific qubit. The most common gate is the $R_y$ rotation:$$R_y(x_i) = \begin{pmatrix} \cos(\frac{x_i}{2}) & -\sin(\frac{x_i}{2}) \\ \sin(\frac{x_i}{2}) & \cos(\frac{x_i}{2}) \end{pmatrix}$$Applying this to $n$ qubits creates a Product State:$$|\psi(x)\rangle = \bigotimes_{i=1}^n R_y(x_i)|0\rangle$$

*Insight:* Note that this mapping is periodic. If your data is not scaled between $[-\pi, \pi]$, the quantum state will "wrap around," causing the model to lose the ability to distinguish between different values.

## III. Stage 2: The Variational Ansatz (The Unitary)
The Ansatz is the "quantum neural network" part of the circuit. It consists of a parameterized unitary $U(\theta)$ composed of gates with tunable parameters $\theta$ (usually rotation angles).

A robust Ansatz must balance **expressibility** (the ability to reach many states in Hilbert space) and entrainability (the ease of optimizing the parameters).

The **Real Amplitudes Ansatz** is a popular choice for heuristic QML. It consists of alternating layers of single-qubit Y-rotations and entangling gates (like CNOTs):$$U(\theta) = \prod_{l=1}^{L} \left( \text{Entanglement} \cdot \bigotimes_{i=1}^{n} R_y(\theta_{i,l}) \right)$$The goal of this unitary is to rotate the state $|\psi(x)\rangle$ into a new state $|\psi(x, \theta)\rangle$ that, when measured, reveals the underlying pattern in the data.

Think of the Ansatz as the "brain" of the QML model. It consists of a parameterized unitary $U(\theta)$ that we tune during training.
### The Requirement of Unitarity
A quantum circuit must be Unitary ($U^\dagger U = I$). This ensures:
* Reversibility: Information is not lost during the computation.
* Norm Preservation: The total probability remains 1.0.

### The Role of Entanglement
To find non-linear correlations, we use entangling gates (like CNOT). Without entanglement, the qubits remain independent, and the model's "Expressibility" is limited to simple linear relationships.

## IV. Stage 3: Observables (The Readout)
After the unitary transformations, we extract classical information by measuring the expectation value of an **Observable** $\hat{M}$. An observable is a Hermitian operator that represents a physical quantity.The expected value $y$ for a given input $x$ and parameters $\theta$ is:$$\langle \hat{M} \rangle = \langle \psi(x, \theta) | \hat{M} | \psi(x, \theta) \rangle$$In binary classification, we often measure the Pauli-Z operator on the first qubit:$$\hat{M} = Z \otimes I \otimes \dots \otimes I$$The output will be a value in the range $[-1, 1]$. We then map this back to a probability or a class label.

Therefore, the observable is a Hermitian operator $\hat{M}$. We extract information by calculating the Expectation Value:$$y = \langle \psi(x, \theta) | \hat{M} | \psi(x, \theta) \rangle$$

### Why Hermitian?
Hermitian operators ($M = M^\dagger$) guarantee that the eigenvalues—the values we actually measure—are real numbers. In binary classification, we typically measure the Pauli-Z operator on a readout qubit:
* Output +1: The state is $|0\rangle$ (Class A).
* Output -1: The state is $|1\rangle$ (Class B).
* Output 0: The state is in an equal superposition (Uncertain).

## V. Hands-on Implementation (Google Colab)

````python
# !pip install amazon-braket-sdk matplotlib -q

import matplotlib.pyplot as plt
from braket.circuits import Circuit, FreeParameter, Observable
from braket.devices import LocalSimulator
import numpy as np

# --- STEP 1: STATE PREPARATION (ANGLE ENCODING) ---
def feature_map(n_qubits):
    qc = Circuit()
    # We create parameters x_0, x_1, etc.
    for i in range(n_qubits):
        phi = FreeParameter(f"x_{i}")
        qc.ry(i, phi) # Ry(angle) on qubit i
    return qc

# --- STEP 2: VARIATIONAL ANSATZ (UNITARY) ---
def ansatz(n_qubits):
    qc = Circuit()
    # Parameters for the trainable weights theta
    for i in range(n_qubits):
        theta = FreeParameter(f"theta_{i}")
        qc.ry(i, theta)

    # Entanglement Layer (CNOTs)
    if n_qubits > 1:
        for i in range(n_qubits - 1):
            qc.cnot(i, i + 1)
    return qc

# --- STEP 3: EXECUTION & OBSERVABLE ---
def run_braket_qml(data_values, weight_values):
    n = len(data_values)

    # Compose the full circuit
    full_circuit = feature_map(n).add_circuit(ansatz(n))

    # Add the expectation value request (Pauli-Z on qubit 0)
    full_circuit.expectation(observable=Observable.Z(), target=0)

    # Define the device (Local Simulator)
    device = LocalSimulator()

    # Create the parameter mapping
    inputs = {f"x_{i}": data_values[i] for i in range(n)}
    weights = {f"theta_{i}": weight_values[i] for i in range(n)}
    all_params = {**inputs, **weights}

    # Run the circuit
    task = device.run(full_circuit, shots=0, inputs=all_params) # shots=0 for exact expectation
    result = task.result()

    return result.values[0], full_circuit

# --- TEST RUN ---
test_x = [np.pi/4, 0.5]
test_theta = [0.2, -0.1]
val, circuit_structure = run_braket_qml(test_x, test_theta)

print("AWS Braket Circuit Structure:")
print(circuit_structure)
print(f"\nModel Output (Expectation <Z>): {val:.4f}")
````

## VI Deeper Understanding of State Preparation
### 1. The Mathematical Objective: Feature Mapping
The core goal is to define a map $\Phi: \mathcal{X} \to \mathcal{H}$ that takes a classical input $x$ and encodes it into a quantum state $|\psi(x)\rangle$.Mathematically, this is expressed as:$$|\psi(x)\rangle = \sum_{i} c_i(x) |i\rangle$$where $c_i(x)$ are complex amplitudes and $\sum_i |c_i(x)|^2 = 1$. In QML, we achieve this by applying a data-dependent unitary $U_{\Phi}(x)$ to the "ground state" $|0\rangle^{\otimes n}$.

### 2. Angle Encoding (Linear Mapping)
In the Python script provided earlier, we used Angle Encoding. This is the most straightforward method because it maps each feature to the rotation angle of a qubit.
### The Unitary:
$$U_{\Phi}(x) = \bigotimes_{i=1}^n R_y(x_i) = \bigotimes_{i=1}^n \begin{pmatrix} \cos(\frac{x_i}{2}) & -\sin(\frac{x_i}{2}) \\ \sin(\frac{x_i}{2}) & \cos(\frac{x_i}{2}) \end{pmatrix}$$

### The Resulting State:
If we have two features $x_1$ and $x_2$, the state becomes:$$|\psi(x)\rangle = (\cos\frac{x_1}{2}|0\rangle + \sin\frac{x_1}{2}|1\rangle) \otimes (\cos\frac{x_2}{2}|0\rangle + \sin\frac{x_2}{2}|1\rangle)$$
* Fundamental Insight: This maps the data into a product state. Because there are no entangling gates, the qubits are independent. This represents a linear combination of basis states.

### 3. The ZZ-Feature Map (Non-Linear & Entangled)
To gain a "quantum advantage" or capture complex patterns, we use non-linear feature maps like the ZZ-Feature Map. This map introduces a "bias" or "interaction" between features $x_1$ and $x_2$.
### The Mathematical Structure:
$$U_{\Phi}(x) = \exp\left(i \phi_{\{1,2\}}(x_1, x_2) Z_1 \otimes Z_2\right) \exp\left(i \phi_1(x_1) Z_1\right) \exp\left(i \phi_2(x_2) Z_2\right)$$Where $\phi_S(x)$ is often a non-linear function like $(\pi - x_i)(\pi - x_j)$.
* Fundamental Insight: The $Z \otimes Z$ interaction creates entanglement. Mathematically, the features are now "correlated" within the quantum state. This is analogous to the Kernel Trick in classical Support Vector Machines (SVMs), where data is projected into a higher-dimensional space to make it linearly separable.

### 4. Why Hilbert Space Matters
When we encode $n$ features into $n$ qubits, we are moving from a space of $n$ dimensions to a space of $2^n$ dimensions.
* Classical Space: $x = [x_1, x_2, \dots, x_n]$ (linear growth).
* Quantum Space: $|\psi(x)\rangle$ is a vector with $2^n$ complex coefficients (exponential growth).

The "fundamental" of QML is that we perform our "learning" (the Ansatz $U(\theta)$) in this massive $2^n$ dimensional Hilbert space, allowing the model to recognize patterns that are mathematically "invisible" in the original $n$-dimensional classical space.

## VII Deeper Understanding of Ansatz
To understand the Ansatz (plural: Ansätze) is to understand the core "learning" engine of a Variational Quantum Algorithm. While the Feature Map prepares the data, the Ansatz is the parameterized model that we tune to find a solution.
Mathematically, an Ansatz is a **parameterized unitary transformation** $U(\theta)$ that explores a specific subspace of the vast Hilbert space.
### 1. The Mathematical Objective: Expressibility vs. Entrainability
When designing or choosing an Ansatz, there is a fundamental trade-off between two mathematical properties:
* **Expressibility**: The ability of the circuit to reach a wide variety of states in the Hilbert space. If your Ansatz is not expressible enough, the "correct" solution may lie in a region of Hilbert space your circuit simply cannot reach.
* **Entrainability**: The ease with which we can optimize the parameters $\theta$. If a circuit is too expressible (highly complex/random), it can lead to Barren Plateaus, where the gradient of the cost function becomes exponentially small, making training impossible.

### 2. The Structure: Hardware-Efficient Ansatz
In the current era of NISQ (Noisy Intermediate-Scale Quantum) devices, we often use **Hardware-Efficient Ansätze**. These are designed to minimize the gate count and depth to avoid noise.

A typical layer in a Hardware-Efficient Ansatz consists of:
* Single-Qubit Rotation Layer: Usually $R_y(\theta)$ or $R_x(\theta)$. These provide the degrees of freedom (the "weights").
* Entanglement Layer: Usually a series of CNOT or CZ gates.

Mathematical Representation of a single layer $l$:$$U_l(\theta) = \left( \prod_{i < j} \text{CNOT}_{i,j} \right) \left( \bigotimes_{k=1}^n R_y(\theta_k) \right)$$

### 3. The Role of Entanglement (The Correlator)
The CNOT gates in the Ansatz are not just "connections"; they are mathematical correlators.
* Without Entanglement: The state $|\psi\rangle$ is a product state. The probability of measuring qubit 1 is mathematically independent of qubit 2. This is equivalent to a classical model that cannot learn interactions between features (e.g., it could learn "Height" and "Weight" separately, but not "BMI").
* With Entanglement: The state becomes non-separable. The unitary $U(\theta)$ can now map the input to states where the value of one qubit is entangled with another. This allows the model to capture high-order, non-linear correlations that are difficult for classical systems to process.

### 4. Specialized Ansätze: Problem-Inspired
In some fields, we don't use generic layers. We use the physics of the problem to guide the math:
* UCCSD (Unitary Coupled Cluster): Used in Quantum Chemistry. The structure of the unitary is derived from electron excitation operators to ensure the circuit stays within a physically valid subspace (e.g., preserving particle number).
* QAOA (Quantum Approximate Optimization Algorithm): Used for optimization. The Ansatz alternates between a "cost Hamiltonian" (representing the problem) and a "mixing Hamiltonian" (allowing the state to explore new solutions).

## VIII Deeper Understanding of Observables
In Quantum Machine Learning, Observables are the bridge between the quantum world of probabilities and the classical world of data. While the state preparation and ansatz manipulate complex probability amplitudes, the observable is what allows us to extract a real-numbered result.
### 1. The Mathematical Definition
In quantum mechanics, an observable is represented by a Hermitian operator $\hat{M}$. A matrix is Hermitian if it is equal to its own conjugate transpose:$$\hat{M} = \hat{M}^\dagger$$
**Why must it be Hermitian?** The Spectral Theorem states that the eigenvalues of a Hermitian operator are always real numbers. Since we use these values to represent physical measurements or class labels (like -1 and 1), the math must guarantee we don't get a complex number as an output.

### 2. The Expectation Value (The "Output")
When we "measure" a quantum circuit in QML, we are usually calculating the Expectation Value. This is the average value you would get if you ran the same circuit and measured it thousands of times.The expectation value $y$ for a state $|\psi\rangle$ is:$$\langle \hat{M} \rangle = \langle \psi | \hat{M} | \psi \rangle$$

### 3. Common Observables in QML
The most frequent choice for binary classification is the Pauli-Z operator.
* Single Qubit Measurement: We measure the first qubit using $\hat{M} = Z$.
  * If the state is $|0\rangle$, the output is $+1$.
  * If the state is $|1\rangle$, the output is $-1$.
  * If the state is a superposition, the output is a value between $-1$ and $1$.
* Multi-Qubit Measurement (Hamiltonians): Sometimes we define a joint observable across multiple qubits. For example, $Z \otimes Z$ measures the parity (correlation) between two qubits.
  * If qubits are the same ($|00\rangle$ or $|11\rangle$), the output is $+1$.
  * If qubits are different ($|01\rangle$ or $|10\rangle$), the output is $-1$.

### 4. Role in the Cost Function
In a variational circuit, the observable output $y(x, \theta)$ is used to calculate the Loss.For example, if we use Mean Squared Error (MSE):$$\mathcal{L}(\theta) = \frac{1}{N} \sum_{i=1}^N (y_{true} - \langle \psi(x_i, \theta) | \hat{M} | \psi(x_i, \theta) \rangle)^2$$The optimization loop then calculates the gradient of this expectation value with respect to the ansatz parameters ($\theta$) to update the weights.

## 3. The "Iris" Workbook (AWS Braket)
This script loads the Iris dataset, prepares a 4-qubit circuit, and defines the training loop structure.

````python
# !pip install amazon-braket-sdk scikit-learn matplotlib -q

import numpy as np
from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler
from braket.circuits import Circuit, FreeParameter, Observable
from braket.devices import LocalSimulator

# 1. DATA PREPARATION
iris = load_iris()
X_data = iris.data # 4 features
y_data = iris.target # 3 classes

# Scaling: Angles must be between 0 and pi for Ry encoding
scaler = MinMaxScaler(feature_range=(0, np.pi))
X_scaled = scaler.fit_transform(X_data)

# 2. STATE PREPARATION (THE CHALLENGE)
# Students must decide: Simple Ry or a more complex ZZ-style map?
def iris_state_prep(n_qubits):
    qc = Circuit()
    for i in range(n_qubits):
        x_param = FreeParameter(f"x_{i}")
        qc.ry(i, x_param)
    # HARD MODE: Add entanglement here to capture feature correlations
    # qc.cnot(0,1).cnot(1,2).cnot(2,3)
    return qc

# 3. THE ANSATZ (THE WEIGHTED UNITARY)
def iris_ansatz(n_qubits, layers=2):
    qc = Circuit()
    for l in range(layers):
        for i in range(n_qubits):
            theta = FreeParameter(f"theta_{l}_{i}")
            qc.ry(i, theta)

        # Hardware-efficient entanglement (Circular CNOTs)
        for i in range(n_qubits):
            qc.cnot(i, (i + 1) % n_qubits)
    return qc

# 4. MEASUREMENT & INFERENCE
def predict(data_point, weights):
    n = len(data_point)
    full_circ = iris_state_prep(n).add_circuit(iris_ansatz(n))

    # We measure Z on the first 2 qubits to get enough "bitstrings" for 3 classes
    full_circ.expectation(Observable.Z(), target=0)
    full_circ.expectation(Observable.Z(), target=1)

    device = LocalSimulator()

    # Map parameters to values
    param_dict = {f"x_{i}": data_point[i] for i in range(n)}
    # Note: Training would involve updating these 'weights' via an optimizer
    weight_dict = {f"theta_{l}_{i}": weights[l][i] for l in range(2) for i in range(n)}

    task = device.run(full_circ, shots=0, inputs={**param_dict, **weight_dict})
    return task.result().values

# Run a test prediction
initial_weights = np.random.uniform(0, np.pi, (2, 4))
result = predict(X_scaled[0], initial_weights)
print(f"Expectation Values (Q0, Q1): {result}")
````

### Challenge 1: The "Non-Linear" Wall
The Problem: Simple $R_y$ encoding maps data linearly. For complex datasets like Iris, the classes are not linearly separable in the original feature space.

The Solution: You must introduce Entangled Feature Maps (like the ZZ-Feature Map). This creates interactions between features $x_i$ and $x_j$, effectively creating a "Quantum Kernel.

"Revised Code for iris_state_prep:

````python
def iris_state_prep_hard(n_qubits):
    qc = Circuit()
    # 1. Single qubit rotations (Linear)
    for i in range(n_qubits):
        qc.ry(i, FreeParameter(f"x_{i}"))

    # 2. Entangling interactions (Non-linear cross-terms)
    # This maps interactions between adjacent features
    for i in range(n_qubits - 1):
        qc.cnot(i, i+1)
        # The non-linear "kernel" term: phi(x_i, x_{i+1})
        # Usually (pi - x_i)(pi - x_{i+1})
        qc.rz(i+1, FreeParameter(f"x_{i}_{i+1}"))
        qc.cnot(i, i+1)
    return qc
````

Mathematical Insight: By adding the $CNOT \to R_z \to CNOT$ structure, you are performing a transformation that cannot be easily replicated by a linear classical model. You are projecting the data into a Hilbert space where the relationship between features is captured in the quantum phase.

### Challenge 3: Barren Plateaus
The Problem: Why do expectation values move toward 0 as layers increase?

The Solution: This is the Barren Plateau phenomenon. As a variational circuit becomes deeper and more "expressive," it begins to resemble a 2-design (a state that is effectively random).

Theory:
In high-dimensional Hilbert spaces, the volume is massive. If your Ansatz $U(\theta)$ is too deep and randomly initialized, the gradient of the cost function $C$:$$\text{Var}[\partial_\theta C] \approx e^{-n}$$where $n$ is the number of qubits.

Mathematical Insight: The variance of the gradient vanishes exponentially. This means the "landscape" of your cost function becomes a vast, flat plain with a single, tiny "needle" of a minimum. You can't "feel" the slope, so the optimizer gets stuck at 0 (the center of the Bloch sphere).

How to fix it:
* 1.  Identity Initialization: Start with weights that make the Ansatz an Identity matrix.
* 2.  Layer-wise training: Train one layer at a time.
* 3.  Local Observables: Measure single qubits rather than global correlations.
