# Let's Review Linear Algebra
To understand Quantum Machine Learning (QML), we must master the "language" of the quantum state: Linear Algebra. Here is a breakdown of the fundamental operations that transform classical vectors into quantum amplitudes.

## 1. Vectors and the "Ket" Notation
In quantum computing, we represent a qubit as a vector in a 2-dimensional complex space. We use Dirac Notation (Bra-Ket):
* Ket $|v\rangle$: A column vector.
* Bra $\langle v|$: The conjugate transpose (row vector) of the ket.

## 2. Inner Product (Scalar Product)
The inner product takes two vectors and returns a single complex number (a scalar). In quantum mechanics, it represents the overlap or similarity between two states.

Notation: $\langle \phi | \psi \rangle$

Calculation: If $|\psi\rangle = \begin{pmatrix} a \\ b \end{pmatrix}$ and $|\phi\rangle = \begin{pmatrix} c \\ d \end{pmatrix}$, then:$$\langle \phi | \psi \rangle = (c^* \cdot a) + (d^* \cdot b)$$
* If $\langle \phi | \psi \rangle = 0$, the states are orthogonal (completely different).
* If $\langle \phi | \psi \rangle = 1$, the states are identical.

## 3. Outer Product (Operator Construction)
The outer product takes two vectors and returns a matrix. In QML, this is how we build "operators" or gates.

Notation: $|\psi\rangle \langle \phi |$

Calculation:$$\begin{pmatrix} a \\ b \end{pmatrix} \begin{pmatrix} c^* & d^* \end{pmatrix} = \begin{pmatrix} ac^* & ad^* \\ bc^* & bd^* \end{pmatrix}$$For example, the project operator $|0\rangle \langle 0|$ is the matrix $\begin{pmatrix} 1 & 0 \\ 0 & 0 \end{pmatrix}$, which "filters" a state to see if it is in the $|0\rangle$ position.

## 4. The Tensor Product ($\otimes$)
The tensor product is the most important operation for scaling. It allows us to combine individual qubits into a single multi-qubit system. If you have two qubits, their combined state space is $2 \times 2 = 4$ dimensions.

Calculation:$$\begin{pmatrix} a \\ b \end{pmatrix} \otimes \begin{pmatrix} c \\ d \end{pmatrix} = \begin{pmatrix} a \cdot c \\ a \cdot d \\ b \cdot c \\ b \cdot d \end{pmatrix}$$

This is how we represent the state $|01\rangle$ by calculating $|0\rangle \otimes |1\rangle$.

## 5. Sum of Products and Sum of Tensors Notation
When describing complex quantum systems or operators, we use summation notation to simplify large matrices.

### Sum of Products (Spectral Decomposition)

Any Hermitian operator (like an observable $M$) can be written as a sum of its eigenvalues ($\lambda_i$) and its eigenvectors ($|v_i\rangle$):$$M = \sum_{i} \lambda_i |v_i\rangle \langle v_i|$$This notation tells us that the operator $M$ is composed of a set of projections onto its various possible measurement outcomes.

### Sum of Tensors (Kronecker Sum)
In QML, we often describe a "Hamiltonian" or a cost function as a sum of interactions between qubits.$$H = \sum_{i,j} w_{ij} (Z_i \otimes Z_j)$$
* $Z_i \otimes Z_j$: This represents a Pauli-Z interaction between the $i$-th and $j$-th qubit.
* $\sum$: This notation indicates that the total energy (or cost) of the system is the cumulative sum of all these individual qubit interactions.

| Operation | Notation | Result Type | QML Purpose |
| :--- | :--- | :--- | :--- |
| **Inner Product** | $\langle \phi \vert \psi \rangle$ | Scalar | **Similarity:** Used to calculate probabilities, overlaps between states, and kernels. |
| **Outer Product** | $\vert \psi \rangle \langle \phi \vert$ | Matrix | **Operators:** Used to construct quantum gates (Unitaries) and measurement observables (Hermitians). |
| **Tensor Product** | $\vert \psi \rangle \otimes \vert \phi \rangle$ | Large Vector | **Scaling:** Used to describe the state of multiple qubits simultaneously ($2^n$ dimensions). |
| **Summation** | $\sum \alpha_i \vert \psi_i \rangle$ | Vector / Matrix | **Superposition:** Describing a state as a weighted sum of basis states or a total Hamiltonian energy. |

## 1. The Inner Product (Similarity)
The inner product $\langle \phi | \psi \rangle$ measures how much one state "overlaps" with another. In QML, this is used to calculate the probability of a specific outcome or to define a "Quantum Kernel."

The Math:

Given $|\psi\rangle = \begin{pmatrix} a \\ b \end{pmatrix}$ and $|\phi\rangle = \begin{pmatrix} c \\ d \end{pmatrix}$:$$\langle \phi | \psi \rangle = (c^* \cdot a) + (d^* \cdot b)$$Python Implementation:

```python
import numpy as np

# Define two states: |0> and |+>
ket0 = np.array([[1], [0]])
ket_plus = (1/np.sqrt(2)) * np.array([[1], [1]])

# Calculate Inner Product <+|0>
# Use np.vdot for complex vectors (automatically conjugates the first vector)
overlap = np.vdot(ket_plus, ket0)

print(f"The overlap <+|0> is: {overlap:.4f}")
print(f"Probability of measuring |+> as |0>: {np.abs(overlap)**2:.4f}")
```

## 2. The Outer Product (Operators)
While the inner product results in a single number, the outer product $|\psi\rangle \langle \phi |$ results in a matrix. This is how we build quantum gates and observables.

The Math:$$|0\rangle\langle 0| = \begin{pmatrix} 1 \\ 0 \end{pmatrix} \begin{pmatrix} 1 & 0 \end{pmatrix} = \begin{pmatrix} 1 & 0 \\ 0 & 0 \end{pmatrix}$$

Python Implementation:Python

```python
# Calculate the outer product |0><0|
projection_0 = np.outer(ket0, ket0.conj().T)

print("The projection matrix |0><0| is:\n", projection_0)

# Verify: Apply this 'filter' to a superposition state
filtered_state = projection_0 @ ket_plus
print("\nApplying |0><0| to |+> results in:\n", filtered_state)
```

## 3. The Tensor Product (Scaling)
The tensor product $\otimes$ is how we describe systems with more than one qubit. It combines two 2-dimensional spaces into one 4-dimensional space.
The Math:

$$|0\rangle \otimes |1\rangle = \begin{pmatrix} 1 \\ 0 \end{pmatrix} \otimes \begin{pmatrix} 0 \\ 1 \end{pmatrix} = \begin{pmatrix} 1 \cdot 0 \\ 1 \cdot 1 \\ 0 \cdot 0 \\ 0 \cdot 1 \end{pmatrix} = \begin{pmatrix} 0 \\ 1 \\ 0 \\ 0 \end{pmatrix}$$

Python Implementation:

```python
# Create the state |01>
ket1 = np.array([[0], [1]])
ket01 = np.kron(ket0, ket1)

print("The 4-dimensional vector for |01> is:\n", ket01)

# Tensor product of gates: X ⊗ I
X_gate = np.array([[0, 1], [1, 0]])
I_gate = np.eye(2)
XI_gate = np.kron(X_gate, I_gate)

print("\nThe 4x4 matrix for X ⊗ I is:\n", XI_gate)
```

## 4. Summation Notation (Interactions)
In QML, complex models are often written as a sum of interactions. For example, a cost function might be the sum of $Z$ measurements across all qubits.

The Math:

$$O = \sum_{i=1}^n Z_i = (Z \otimes I \otimes I \dots) + (I \otimes Z \otimes I \dots) + \dots$$Python Implementation:

```python
# Calculate the total Z-energy for a 2-qubit system: Z0 + Z1
Z = np.array([[1, 0], [0, -1]])
I = np.eye(2)

# Z0 = Z ⊗ I
Z0 = np.kron(Z, I)
# Z1 = I ⊗ Z
Z1 = np.kron(I, Z)

Total_Z = Z0 + Z1
print("The Sum of Tensors (Z0 + Z1) is:\n", Total_Z)
```

# Applying to QML
We will transition from static matrix multiplication into the dynamic evolution of a quantum circuit. We will work through the math of a single gate, the composition of two gates, and finally the Quantum Fourier Transform (QFT), which is the "square root of a permutation" and a cornerstone of quantum algorithms.

## 1. Single Gate Evolution: The Hadamard ($H$)
The Hadamard gate is the most common way to create superposition. It maps the basis states $|0\rangle$ and $|1\rangle$ to the "equator" of the Bloch sphere.

The Math:
$$H = \frac{1}{\sqrt{2}}\begin{pmatrix} 1 & 1 \\ 1 & -1 \end{pmatrix}$$Applying $H$ to $|0\rangle$:$$H|0\rangle = \frac{1}{\sqrt{2}}\begin{pmatrix} 1 & 1 \\ 1 & -1 \end{pmatrix} \begin{pmatrix} 1 \\ 0 \end{pmatrix} = \frac{1}{\sqrt{2}}\begin{pmatrix} 1 \\ 1 \end{pmatrix} = |+\rangle$$

Braket Implementation:Python

```python
from braket.circuits import Circuit
from braket.devices import LocalSimulator

qc = Circuit().h(0)
device = LocalSimulator()
# shots=0 gives the exact state vector
result = device.run(qc, shots=0).result()
print("State vector after H gate:\n", result.values[0])
```

## 2. Two-Gate Composition: $Z$ after $H$
When we apply multiple gates, we multiply the matrices in reverse order of the circuit diagram. To apply $H$ then $Z$, the math is $Z \times H$.

The Math:

$$ZH|0\rangle = \begin{pmatrix} 1 & 0 \\ 0 & -1 \end{pmatrix} \left[ \frac{1}{\sqrt{2}}\begin{pmatrix} 1 \\ 1 \end{pmatrix} \right] = \frac{1}{\sqrt{2}}\begin{pmatrix} 1 \\ -1 \end{pmatrix} = |-\rangle$$The $Z$ gate flipped the phase of the $|1\rangle$ component.

Braket Implementation:

```python
qc = Circuit().h(0).z(0)
result = device.run(qc, shots=0).result()
print("State vector after H and Z gates:\n", result.values[0])
```

## 3. The 2-Qubit Quantum Fourier Transform (QFT)
The QFT performs a Fourier transform on the amplitudes of a quantum state. For 2 qubits, the circuit involves Hadamard gates and a controlled-phase rotation ($R_z$ or $CPHASE$).
The 2-Qubit QFT Circuit Logic:
* Apply $H$ to Qubit 0.
* Apply Controlled-Phase ($R_z$) between Qubit 0 and Qubit 1.
* Apply $H$ to Qubit 1.
* Swap the qubits (optional, depending on convention).

The Math (The Operator):

The QFT matrix for $N$ dimensions (where $N=2^n$) is defined by the root of unity $\omega_N = e^{2\pi i / N}$. For 2 qubits ($N=4$), $\omega_4 = e^{i\pi/2} = i$.$$QFT_4 = \frac{1}{2} \begin{pmatrix} 1 & 1 & 1 & 1 \\ 1 & i & -1 & -i \\ 1 & -1 & 1 & -1 \\ 1 & -i & -1 & i \end{pmatrix}$$

Braket Implementation:

```python
import numpy as np

# QFT Circuit for 2 qubits
qft_circ = Circuit()
qft_circ.h(0)
# CPhase(pi/2)
qft_circ.cphaseshift(1, 0, np.pi/2)
qft_circ.h(1)
# Swap qubits to match standard QFT ordering
qft_circ.swap(0, 1)

print("2-Qubit QFT Circuit:")
print(qft_circ)

# Run on |00>
result = device.run(qft_circ, shots=0).result()
print("\nFinal State Vector (Should be equal superposition [0.5, 0.5, 0.5, 0.5]):")
print(result.values[0])
```

To derive the output of a 2-qubit Quantum Fourier Transform (QFT), we must move from the abstract circuit diagram to the specific matrix transformations applied to the state vector.The general formula for a QFT on $n$ qubits (where $N = 2^n$) mapping a state $|j\rangle$ is:$$QFT|j\rangle = \frac{1}{\sqrt{N}} \sum_{k=0}^{N-1} e^{2\pi i j k / N} |k\rangle$$For 2 qubits ($N=4$), let’s derive the output for the input state $|j\rangle = |1\rangle$ (which is the binary state $|01\rangle$).

### The Step-by-Step Circuit Derivation
We start with the state $|\psi_0\rangle = |01\rangle = |0\rangle \otimes |1\rangle$.

### Step A: Hadamard on Qubit 0
The Hadamard transform $H$ maps $|0\rangle \to \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle)$.$$|\psi_1\rangle = (H \otimes I)|01\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle) \otimes |1\rangle = \frac{1}{\sqrt{2}}(|01\rangle + |11\rangle)$$

### Step B: Controlled-Phase Shift ($CPHASE$)
The $CPHASE$ gate applies a phase $e^{i\theta}$ only if both qubits are $|1\rangle$. For QFT, the angle is $\theta = 2\pi/2^2 = \pi/2$. Note that $e^{i\pi/2} = i$.
* $|01\rangle$ remains $|01\rangle$.
* $|11\rangle$ becomes $e^{i\pi/2}|11\rangle = i|11\rangle$.$$|\psi_2\rangle = \frac{1}{\sqrt{2}}(|01\rangle + i|11\rangle)$$

### Step C: Hadamard on Qubit 1
Now we apply $H$ to the second qubit. Remember $H|1\rangle = \frac{1}{\sqrt{2}}(|0\rangle - |1\rangle)$.$$|\psi_3\rangle = \frac{1}{\sqrt{2}} \left( |0\rangle \otimes \frac{|0\rangle - |1\rangle}{\sqrt{2}} + i |1\rangle \otimes \frac{|0\rangle - |1\rangle}{\sqrt{2}} \right)$$

Multiply the terms out:$$|\psi_3\rangle = \frac{1}{2} \left( |00\rangle - |01\rangle + i|10\rangle - i|11\rangle \right)$$

### Step D: The Swap Gate
Standard QFT convention requires a swap at the end to correct bit-ordering ($|q_0 q_1\rangle \to |q_1 q_0\rangle$).$$|\psi_{final}\rangle = \frac{1}{2} \left( |00\rangle + i|01\rangle - |10\rangle - i|11\rangle \right)$$

### Deriving the Measurement Output (Expectation Values)
In QML, we don't just look at the state vector; we measure an Observable, typically the Pauli-Z operator ($Z$).

If we measure $Z$ on the first qubit ($Z \otimes I$), the expectation value is calculated as $\langle \psi | Z \otimes I | \psi \rangle$.

The Math:
Apply the Operator: $Z$ keeps $|0\rangle$ terms positive and makes $|1\rangle$ terms negative.$$(Z \otimes I)|\psi_{final}\rangle = \frac{1}{2} \left( |00\rangle + i|01\rangle - (-|10\rangle) - (-i|11\rangle) \right) = \frac{1}{2} \left( |00\rangle + i|01\rangle + |10\rangle + i|11\rangle \right)$$

Inner Product: Multiply the original "bra" $\langle \psi_{final} |$ by this new vector:$$\langle Z_0 \rangle = \frac{1}{4} \left( (1)(1) + (-i)(i) + (-1)(1) + (i)(i) \right)$$$$\langle Z_0 \rangle = \frac{1}{4} \left( 1 + 1 - 1 - 1 \right) = 0$$

In quantum mechanics, there is a fundamental difference between how we draw a circuit and how we calculate it. This distinction is the most common source of error for students.

### The Visual vs. Algebraic Conflict
* Visual Flow (Left to Right): When you look at a circuit diagram, you read it like a sentence. Time flows from left to right. Qubit 0 starts at the far left, passes through Gate A, then Gate B, and finally Gate C.
* Algebraic Flow (Right to Left): In linear algebra, operators are applied to a vector from the right. If a state $|\psi\rangle$ is acted upon by Gate A, then Gate B, and then Gate C, the equation is written as:$$|\psi_{final}\rangle = \mathbf{C} \cdot \mathbf{B} \cdot \mathbf{A} \cdot |\psi_{initial}\rangle$$

### Why do we go "Right to Left"?
The reason is rooted in the definition of a function. In standard math, if you have a function $f(x)$ and you want to apply $g$ to the result, you write $g(f(x))$. The function that happens first is closest to the variable.

In a quantum circuit:
* The state $|\psi\rangle$ is the "variable."
* The gate closest to $|\psi\rangle$ is the one that happens first in time.
* Therefore, the "oldest" gate is on the far left of the equation, and the "newest" gate is on the far right of the diagram.

### Step-by-Step Mathematical Derivation
Let's take a 1-qubit circuit where we apply a Hadamard (H) gate followed by a Pauli-X gate.
The Diagram:

|0> ---[ H ]---[ X ]--->

The Math (Right-to-Left):

We must calculate $X \cdot H \cdot |0\rangle$.

Step A: Apply H (The First Gate)

$$H|0\rangle = \frac{1}{\sqrt{2}} \begin{pmatrix} 1 & 1 \\ 1 & -1 \end{pmatrix} \begin{pmatrix} 1 \\ 0 \end{pmatrix} = \frac{1}{\sqrt{2}} \begin{pmatrix} 1 \\ 1 \end{pmatrix}$$

Step B: Apply X (The Second Gate)

Now we take the result of Step A and multiply it by $X$ from the left.$$X \left( H|0\rangle \right) = \begin{pmatrix} 0 & 1 \\ 1 & 0 \end{pmatrix} \left[ \frac{1}{\sqrt{2}} \begin{pmatrix} 1 \\ 1 \end{pmatrix} \right] = \frac{1}{\sqrt{2}} \begin{pmatrix} 1 \\ 1 \end{pmatrix}$$

### Matrix Composition
If you want to find a single matrix that represents the entire circuit, you must multiply the matrices in reverse-chronological order.

If $U_{total} = X \cdot H$:$$U_{total} = \begin{pmatrix} 0 & 1 \\ 1 & 0 \end{pmatrix} \cdot \frac{1}{\sqrt{2}} \begin{pmatrix} 1 & 1 \\ 1 & -1 \end{pmatrix} = \frac{1}{\sqrt{2}} \begin{pmatrix} 1 & -1 \\ 1 & 1 \end{pmatrix}$$

### Multi-Qubit Order (Tensor Products)
When you have multiple wires, the "Right to Left" rule still applies to the sequence of gates on a wire, but we also have a "Top to Bottom" rule for the Tensor Product.In most frameworks (like AWS Braket), the top wire is Qubit 0 and the bottom wire is Qubit 1. The combined state is $|q_0\rangle \otimes |q_1\rangle$.
* Vertical (Space): Top to Bottom $\to$ $|q_0\rangle \otimes |q_1\rangle \otimes |q_2\rangle$
* Horizontal (Time): Right to Left $\to$ $Gate_3 \cdot Gate_2 \cdot Gate_1$
